```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
from keras.models import Sequential
from keras.layers import Dense, LeakyReLU, ReLU, ELU
```

```
data = pd.read_csv("https://raw.githubusercontent.com/lRAHULl/Deep-Learning/master/Churn_Modelling.csv")
```

```
data
```

```
X = data.iloc[:, 3: -1]
y = data.iloc[:, -1]
```

```
X
```

```
y
```

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```
le1 = LabelEncoder()
X.iloc[:, 1] = le1.fit_transform(X.iloc[:, 1])
le2 = LabelEncoder()
X.iloc[:, 2] = le2.fit_transform(X.iloc[:, 2])
```

```
X
```

```
ohe = OneHotEncoder(categorical_features = [1])
X = ohe.fit_transform(X).toarray()
X
```

```
len(X[1])
```

```
X = X[:, 1:] # to avoid Duplication
len(X[1])
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)
```

```
# Standard Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```
# Define the model
model = Sequential()

# Input Layer
model.add(Dense(units=6, kernel_initializer='uniform', input_dim = 11))
model.add(LeakyReLU())

# Hidden Layer
model.add(Dense(units=20, kernel_initializer='uniform'))
model.add(LeakyReLU())
model.add(Dense(units=20, kernel_initializer='uniform'))
model.add(LeakyReLU())
model.add(Dense(units=20, kernel_initializer='uniform'))
model.add(LeakyReLU())
model.add(Dense(units=20, kernel_initializer='uniform'))
model.add(LeakyReLU())
model.add(Dense(units=20, kernel_initializer='uniform'))
model.add(LeakyReLU())


# Output Layer
model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=100)
```

```
# Prfedicting on the test set
y_pred = model.predict(X_test)
y_pred = y_pred > 0.5
```

```
#  Confusion matrix 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
